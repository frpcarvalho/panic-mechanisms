```python
import pandas as pd
import numpy as np

# ==================== CARREGAR E CRIAR SUBSET ====================
print("üîÑ Carregando dataset original...")
df_full = pd.read_csv('/Users/filipecarvalho/Documents/data_science_projects/Panic.3/NHANES_panic_subset_with_target_modified.csv')

# Definir as 12 features (ordem de import√¢ncia)
features_selected = [
    'BIX_BIDFAT',      # 1. Massa gorda corporal (kg) - 23.8%
    'DEMO_INDFMPIR',   # 2. R√°cio pobreza/rendimento - 17.4%
    'DEMO_RIDAGEMN',   # 3. Idade (meses) - 10.9%
    'DEMO_RIAGENDR',   # 4. G√©nero - 10.6%
    'BPX_BPXDAR',      # 5. Press√£o arterial diast√≥lica - 9.8%
    'DEMO_DMDHHSIZ',   # 7. Tamanho agregado familiar - 7.9%
    'BMX_BMXBMI',      # 8. √çndice Massa Corporal - 7.4%
    'ALQ_ALQ130',      # 9. Consumo √°lcool (drinks/dia) - 2.6%
    'MCQ_MCQ250F',     # 10. Hist√≥ria familiar HTA/AVC - 0.9%
    'MPQ_MPQ070',      # 11. Dor lombar √∫ltimos 3 meses - 0.5%
    'DEMO_DMDMARTL'    # 12. Estado civil - 0.3%
]

# Criar subset
print(f"\nüìä Criando subset com {len(features_selected)} features + target...")
df_subset = df_full[features_selected + ['target']].copy()

# ==================== INFORMA√á√ïES B√ÅSICAS ====================
print("\n" + "="*80)
print("‚úÖ SUBSET CRIADO COM SUCESSO")
print("="*80)

print(f"\nüî¢ Shape do subset: {df_subset.shape[0]} linhas √ó {df_subset.shape[1]} colunas")
print(f"üíæ Redu√ß√£o: {df_full.shape[1]} ‚Üí {df_subset.shape[1]} colunas ({df_full.shape[1] - df_subset.shape[1]} colunas removidas)")

print("\nüìã COLUNAS NO SUBSET:")
for i, col in enumerate(df_subset.columns, 1):
    print(f"  {i:2d}. {col}")

print("\nüéØ DISTRIBUI√á√ÉO DO TARGET:")
target_counts = df_subset['target'].value_counts().sort_index()
for val, count in target_counts.items():
    label = "Normal" if val == 0 else "Panic Disorder"
    pct = (count / len(df_subset)) * 100
    print(f"  {label:15s} (target={val}): {count:5d} casos ({pct:5.2f}%)")

print("\nüìä MISSING VALUES POR FEATURE:")
missing = df_subset.isnull().sum()
missing_pct = (missing / len(df_subset)) * 100
for feat in features_selected:
    if missing[feat] > 0:
        print(f"  ‚ö†Ô∏è  {feat:20s}: {missing[feat]:4d} missing ({missing_pct[feat]:5.2f}%)")
    else:
        print(f"  ‚úÖ {feat:20s}: 0 missing")

print("\nüìà ESTAT√çSTICAS DESCRITIVAS B√ÅSICAS:")
print(df_subset[features_selected].describe().round(2))

# ==================== SALVAR SUBSET ====================
output_path = '/Users/filipecarvalho/Documents/data_science_projects/Panic.3/NHANES_panic_11features.csv'
print(f"\nüíæ Salvando subset em:")
print(f"   {output_path}")
df_subset.to_csv(output_path, index=False)
print("   ‚úÖ Salvo com sucesso!")

print("\n" + "="*80)
print("üéâ PRONTO PARA AN√ÅLISES!")
print("="*80)
print(f"\nAgora tens um dataset limpo com apenas {len(features_selected)} features + target")
print("Tamanho do arquivo reduzido significativamente para an√°lises mais r√°pidas!")
```
